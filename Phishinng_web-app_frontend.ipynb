```python
%%writefile app.py
import streamlit as st
import joblib
from PIL import Image

# Load model
model_path = "/content/drive/MyDrive/project/model.pkl"

model = joblib.load(model_path)


# Page configuration
st.set_page_config(
    page_title="Phishing Guard App",
    page_icon="🎣",
    layout="centered",
    initial_sidebar_state="expanded",
)

# Define Streamlit app layout
st.title("Phishing Guard App")
st.write("Protect yourself from phishing emails! Enter an email text below and let us check it for you.")

#image
image = Image.open("/content/drive/MyDrive/project/pic.jpg")
st.image(image, use_column_width=True, caption="Image: phishing illustration")

# Input field for user to enter email text
email_text = st.text_area("Enter email text here:")

# Prediction logic
if st.button("Check for Phishing"):
    if email_text.strip() == "":
        st.warning("Please enter some text.")
    else:
        new_data = [email_text]
        pred = model.predict(new_data)
        if pred[0] == "Phishing Email":
            st.error("⚠️ This looks like a phishing attempt! Be cautious.")
        else:
            st.success("✅ This email appears to be safe.")



```
